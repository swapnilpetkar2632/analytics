```python
import matplotlib.pyplot as plt

brands = ["A", "B", "C", "D", "E"]
x = [22, 35, 30, 3, 10]

# explode only the 3rd slice ("C")
explode = [0, 0, 0.1, 0, 0]

plt.pie(
    x,
    labels=brands,
    shadow=True,
    explode=explode,
    normalize=True
)
plt.savefig("bar.jpg")

plt.show()
```
