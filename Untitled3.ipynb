```python
def cheak_prime(num):
    if num==1:
        print("it is prime")
    if num ==  2:
          print("it is prime")
    for i in range(2,num):
        if num % i == 0 :
            print("it is not prime no")
            break
    else:
            print("it is prime no")
cheak_prime(45)
```
