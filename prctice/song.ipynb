```python
data=pd.read_csv("C:\\Users\\Sawpanil\\Downloads\\archive (4)\\Spotify Most Streamed Songs.csv")
```

```python
data=pd.DataFrame(data)
```

```python
data = data.drop(columns=["cover_url"])
```

```python
data
```

```python
data.isnull().sum()
```

```python

data.columns = (
    data.columns
    .str.strip()            # remove leading/trailing spaces
    .str.lower()            # convert to lowercase
    .str.replace(" ", "_")  # replace spaces with underscores
    .str.replace(r"[^\w_]", "", regex=True)  # remove special characters
)
```

```python
data.isnull().sum()
```

```python
data=data.dropna()
```

```python
data.isnull().sum()
```

```python
data
```

```python
data = data.applymap(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
import pandas as pd

# Loop through all object (string) columns
for col in data.select_dtypes(include="object").columns:
    data[col] = data[col].str.replace(r"[^\w_]", "", regex=True)
```
