```python
import pandas as pd

# Load dataset


# Create Service / Plan dimension table
service_table = data[['customerID',
                    'PhoneService',
                    'MultipleLines',
                    'InternetService',
                    'OnlineSecurity',
                    'OnlineBackup',
                    'DeviceProtection',
                    'TechSupport',
                    'StreamingTV',
                    'StreamingMovies',
                    'Contract',
                    'PaperlessBilling',
                    'PaymentMethod']].copy()
service_table.replace({
    'No internet service': 'No',
    'No phone service': 'No'
}, inplace=True)


# Preview
print(service_table.head())

# Save as new CSV file
service_table.to_csv(r"C:\\Users\\Sawpanil\\Downloads\\service_plan_table.csv", index=False)
```

```python
import pandas as pd

# Load dataset


# Create Billing & Usage fact table
billing_table = data[['customerID',
                    'tenure',
                    'MonthlyCharges',
                    'TotalCharges',
                    'Churn']].copy()

# Handle missing TotalCharges (11 null values)
billing_table['TotalCharges'] = billing_table['TotalCharges'].fillna(0)

# Optional: Convert Churn from 0/1 to Yes/No
billing_table['Churn'] = billing_table['Churn'].map({0: 'No', 1: 'Yes'})

# Preview
print(billing_table.head())

# Save as new CSV file
billing_table.to_csv(r"C:\\Users\\Sawpanil\\Downloads\\billing_usage_table.csv", index=False)
```
